```python
#Line plot
x = [1, 2, 3, 4, 5] 
y = [10, 12, 8, 15, 20] 
plt.plot(x, y) 
plt.title("Line Plot") 
plt.xlabel("X-axis") 
plt.ylabel("Y-axis") 
plt.show() 
```

```python
#Bar plot
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.DataFrame({
    'Category': ['A', 'B', 'C'], 
    'Values': [10, 20, 15]
}) 
sns.barplot(x='Category', y='Values', data=data) 
plt.title("Bar Plot") 
plt.show() 
```

```python
#Count plot
sns.countplot(x='Category', data=data) 
plt.title("Count Plot") 
plt.show() 
```

```python
#Histogram
import numpy as np
data = np.random.randn(100) 
plt.hist(data, bins=10) 
plt.title("Histogram") 
plt.show() 
```

```python
#or using seaborn
sns.histplot(data, bins=10, kde=True) 
plt.title("Histogram with KDE") 
plt.show() 
```

```python
#KDE Plot (Kernel Density Estimate) 
sns.kdeplot(data, shade=True) 
plt.title("KDE Plot") 
plt.show() 
```

```python
#Box plot
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(y=data) 
plt.title("Box Plot") 
plt.show() 
```

```python
#Violin Plot
sns.violinplot(y=data) 
plt.title("Violin Plot") 
plt.show() 
```

```python
#Scatter Plot
x = [1, 2, 3, 4, 5] 
y = [5, 4, 7, 10, 8] 
plt.scatter(x, y) 
plt.title("Scatter Plot") 
plt.show() 
```

```python
#Using seaborn
sns.scatterplot(x=x, y=y) 
plt.title("Scatter Plot") 
plt.show() 
```

```python
#Load dataset
df=sns.load_dataset('iris')
df
```

```python
#Pair Plot
df = sns.load_dataset('iris') 
sns.pairplot(df, hue='species') 
plt.show() 
```

```python
#Heatmap
corr = df.corr() 
sns.heatmap(corr, annot=True, cmap='coolwarm') 
plt.title("Correlation Heatmap") 
```

```python
#Pie chart
labels = ['A', 'B', 'C'] 
sizes = [30, 50, 20] 
plt.pie(sizes, labels=labels, autopct='%1.1f%%') 
plt.title("Pie Chart") 
plt.show() 
```

```python
#Swarm plot
sns.swarmplot(x='species', y='sepal_length', data=df) 
plt.title("Swarm Plot") 
plt.show() 
```

```python
#Strip plot
sns.stripplot(x='species', y='sepal_length', data=df, jitter=True) 
plt.title("Strip Plot") 
plt.show() 
```

```python
#Joint plot
sns.jointplot(x='sepal_length', y='petal_length', data=df, kind='scatter') 
plt.show() 
```
