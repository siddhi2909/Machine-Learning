```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import dataset
ds=pd.read_csv("Position_Salaries.csv")
ds
```

```python
##Divide dataset as dependent and independent variable

x=ds[["Level"]]
y=ds[["Salary"]]
```

```python
x
```

```python
y
```

```python
#Polt graph
plt.scatter(x,y)
plt.show()
```

```python
#Model fitting (Polynomial regression model fitting)
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=5)
x_poly=poly_reg.fit_transform(x)
#x_poly
reg=LinearRegression()
reg.fit(x_poly,y)
```

```python
#Visualization
plt.scatter(x,y)
plt.plot(x,reg.predict(poly_reg.transform(x)),color='red')
plt.show()
```

```python
#Predicting for my input
myinput=pd.DataFrame(data=[[11]],columns=["Level"])
result=reg.predict(poly_reg.transform(myinput))
result
```

```python
#Preduct for user input
level=float(input("Enter level:"))
myinput=pd.DataFrame(data=[[level]],columns=["Level"])
poly_input=poly_reg.transform(myinput)
result=reg.predict(poly_input)
print(f"Predicted result for level{level}:{round(result[0,0])}")
```
