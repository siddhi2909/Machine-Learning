```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#Importing and reading dataset
ds=pd.read_csv("Position_Salaries.csv")
ds
```

```python
#Splitting into dependent and independent variable
x=ds[["Level"]]
y=ds[["Salary"]]
```

```python
x
```

```python
y
```

```python
#Fitting Random forest regression model
from sklearn.ensemble import RandomForestRegressor
reg=RandomForestRegressor(n_estimators=10,random_state=0)
reg.fit(x,y)
```

```python
#Predict for my input
reg.predict([[4]])
```

```python
reg.feature_importances_
```

```python
#Visualization of random forest regression tree
from sklearn.tree import plot_tree
plot_tree(reg.estimators_[0],filled=True)
plt.show()
```

```python
plot_tree(reg.estimators_[1],filled=True)
plt.show()
```

```python
#Predict for user input
level=float(input("Enter the level:"))
myinput=pd.DataFrame(data=[[level]],columns=["Level"])
result=reg.predict(myinput)
print(f"Predicted salary for level {level}: ₹{result[0]}")

 
```
