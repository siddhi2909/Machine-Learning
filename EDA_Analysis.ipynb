```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds=pd.read_csv("employee_dataset_for_eda.csv")
print(ds.to_string())
```

```python
#employee count per dept
sns.countplot(data=ds,x='Department',palette='Set2')
plt.title("employee count per dept")
```

```python
#Gender distibution
sns.countplot(data=ds,x='Gender',palette='Set1')
plt.title("Gender distribution")
```

No.of female employees are grater than male employess

```python
#Age distribution
sns.histplot(ds['Age'],bins=10,kde=True)
plt.title('Age distribution')
```

Highest age count is 45 to 50 followed by 25 to 30

```python
#location distribution
sns.countplot(data=ds,x='Location',palette='Spectral')
plt.title("location distribution")
plt.xticks(rotation=45)
```

More employees living in seattle and less employee in Chicago

```python
#Years of experience
sns.histplot(ds['Years_of_Experience'],bins=15,kde=True)
plt.title('Years of experience')
```

Highest no of years of experience of employee is 1.5 to 4.5(no of employee is 15) followed by 10.0 to 12.5(no of employee is 13) 

```python
#Average salry of Department
sns.barplot(data=ds,x='Department',y='Monthly_Salary',estimator=np.mean,palette='viridis')
plt.title("Average salry of Department")
```

Highest salary of department is finance

```python
#Performance score by department
sns.boxplot(data=ds,x='Department',y='Performance_Score',palette='pastel')
plt.title('Performance score by department')
```
