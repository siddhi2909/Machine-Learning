```python
#Handling missing values(using mean and scikit Learn)
from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values=np.nan,strategy='mean')
ds[["Age","Salary"]]=si.fit_transform(ds[["Age","Salary"]])
#fit transform is used to fill the missing values
#simpleImputer for finding single mean 
```

```python
ds
```

```python
#Encoding(OneHotEncoder)
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output=False)
encoded_data=ohe.fit_transform(ds[['Country']])
encoded_data
```

```python
#make a new column for array
column_names=ohe.get_feature_names_out(["Country"])
encoded_df=pd.DataFrame(encoded_data,columns=column_names)
encoded_df
```

```python
#Join it to original dataset
ds=pd.concat([ds,encoded_df],axis=1)
ds
```

```python
#Drop previous columns which are now not necessary
ds=ds.drop(columns=['Country'])
ds
```

```python
#LabelEncoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ds["Purchased"]=le.fit_transform(ds["Purchased"])
ds
```

```python
#Feature scaling (Reduce large data values to it's smaller format)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
ds[['Age','Salary']]=sc.fit_transform(ds[['Age','Salary']])
```

```python
ds
```

```python
#Divide dataset as Independent(input) and Dependent (output) variable
x=ds[['Age','Salary','Country_France','Country_Germany','Country_Spain']]
y=ds[['Purchased']]
```

```python
#Splittting data into Training and Testing dataset

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```
